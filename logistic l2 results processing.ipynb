```python
import ast
import pandas as pd
```

```python
with open('results_logistic_l2_raw.txt', 'r') as f:
    results_lines = f.readlines()
    
results = []

for line in results_lines:
    results.append( ast.literal_eval(line) )
    
for result in results:
    result['features'] = ast.literal_eval(result['features'])
    result['params'] = ast.literal_eval(result['params'])
    
result_tuples = []
for result in results:
    result_tuples.append( (result['error'], result['params']['C'], result['params']['max_iter'] ) )
```

```python
min(results, key=lambda x: x['error'])
```

```python
with open('results_logistic_l2.txt', 'w') as f:
    f.write('error, lambda, num_iter, features\n')
    
    for result in results:
        f.write('{}, {}, {}, {}\n'.format(result['error'], result['params']['C'], result['params']['max_iter'], result['features']))
```

```python
df = pd.DataFrame(result_tuples, columns=['error', 'C', 'max_iter'])
```
